```python
import pandas as pd

df = pd.read_csv('Encoding Data.csv')
df.head()
```

FEATURE ENCODING

1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['bin_1'] = le.fit_transform(df['bin_1'])
df['bin_2'] = le.fit_transform(df['bin_2'])
df.head()
```

2. Binary Encoder

```python
!pip install category_encoders
```

```python
import category_encoders as ce

binary_encoder = ce.BinaryEncoder(cols=['nom_0'])
df_binary = binary_encoder.fit_transform(df[['nom_0']])
df = pd.concat([df, df_binary], axis=1)
df.head()
```

3. Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_order = [['Cold', 'Warm', 'Hot']]
oe = OrdinalEncoder(categories=ordinal_order)

df['ord_2'] = oe.fit_transform(df[['ord_2']])
df.head()
```

4. One-Hot Encoding

```python
df_encoded = pd.get_dummies(df, columns=['nom_0'])
df_encoded.head()
```
